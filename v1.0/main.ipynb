```python
events = get_links(url='http://www.ufcstats.com/statistics/events/completed?page=all')
events.date = pd.to_datetime(events.date)
events = events[events.date < get_current_date()].reset_index(drop=True)
events.tail()
```

```python
#get existing data
try:
    exist_data = pd.read_csv('../data_v1/data_pre_processed.csv',index_col=0)
    exist_events_url = list(np.unique(exist_data.event_url))
except:
    print('Cant find ../data_v1/data_pre_processed.csv')
    exist_data = []
    exist_events_url = []

print(f'Current Existing events: {len(exist_events_url)}')
```

```python
#get new data
new_events = list(set(events.event_url) - set(exist_events_url))
print(f'New events: {len(new_events)}')
```

```python
new_data = []
for url in tqdm(new_events):
    try:
        new_data.extend(get_event_data(url))
    except:
        continue   

data = pd.concat([exist_data,pd.DataFrame(new_data)])
data.tail(3)    
```

```python
pd.DataFrame(data).to_csv('../data_v1/data_pre_processed.csv',index=[0])
```
